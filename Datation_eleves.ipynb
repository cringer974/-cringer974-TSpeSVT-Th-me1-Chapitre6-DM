```python
u=int(input("Valeur de u = ? "))       # Demande d'entrer un nombre entier que l'on stocke dans la variable u
 
def compteur_1():                      # On définit une fonction nommée compteur_1
    while u<50:
        u=u+1                          # on établit un compteur
    return(u)             

print("le résultat pour la fonction compteur_1 est : ", compteur_1())  # affiche une phrase prédéfinie et ce qui est stocké dans la fonction
```

Vous devriez avoir une erreur disant : "*UnboundLocalError: local variable 'u' referenced before assignment*". <br/>
<br/>
La solution consiste à dire à la fonction *compteur_1* d'utiliser une *variable globale*, en tapant *global u*.<br/>
**Etudier** et **exécuter** la cellule ci-dessous.

```python
u=int(input("Valeur de u = ? "))       # Demande d'entrer un nombre entier que l'on stocke dans la variable u
 
def compteur_1():                      # On définit la fonction. les parenthèses vides sont obligtoires
    global u                           # Python recherche u en dehors de l'espace local de la fonction
    while u<50:
        print("A l'intérieur de la fonction, u =",u)  
        u=u+1                           # on établit un compteur
    return(u)          

print("le résultat final de la fonction compteur_1 est : ", compteur_1())  # affiche une phrase prédéfinie et ce qui est stocké dans la fonction
```

La cellule ci-dessus introduit également une autre fonction : la boucle *tant que* (ou *while* en anglais) qui va vérifier un condition (ici que u<50).
Comme vous pouvez le voir, ici, elle permet de créer un compteur dont la valeur augmente automatiquement de 1 
à l'intérieur de la fonction jusqu'à ce que u atteigne 50. A ce moment là, le programme remonte la valeur de u à la fonction dont le résultat s'affiche.<br/>
**Vous pouvez vous amuser à changer les valeurs dans la boucle ou le compteur dans la cellule ci-dessus pour voir ce qui se passe.**
